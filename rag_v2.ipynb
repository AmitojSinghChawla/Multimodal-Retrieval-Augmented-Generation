```python
from unstructured.partition.pdf import partition_pdf

pdf_path = "D:/Projects/Multimodal-Retrieval-Augmented-Generation/uploaded_pdfs/NIPS-2017-attention-is-all-you-need-Paper.pdf"

elements = partition_pdf(
    filename=pdf_path,                  # mandatory
    strategy="hi_res",                                     # mandatory to use ``hi_res`` strategy
    extract_images_in_pdf=True,                            # mandatory to set as ``True``
    extract_image_block_types=["Image"],          # optional
    extract_image_block_to_payload=True,                  # optional
    # extract_image_block_output_dir="D:/Projects/Rag/extracted_images",# optional - only works when ``extract_image_block_to_payload=False``
    )

print(f"Extracted {len(elements)} elements.")


```

```python
# seperate the elements into text, image and table elements
image_elements=[el for el in elements if el["type"] == "image"]
table_elements=[el for el in elements if el[type]== "table"]
text_elements=[el for el in elements if el[type]== "text"]

visual_elements=image_elements + table_elements

from unstructured.chunking.title import chunk_by_title
chunked_texts=chunk_by_title(
    max_characters=10000,
    combine_text_under_n_chars=2000,
    new_after_n_chars=6000,
)

```

```python
import base64
from IPython.display import Image, display

def display_base64_image(base64_code):
    # Decode the base64 string to binary
    image_data = base64.b64decode(base64_code)
    # Display the image
    display(Image(data=image_data))

display_base64_image(visual_elements[1].metadata.image_base64)
```
